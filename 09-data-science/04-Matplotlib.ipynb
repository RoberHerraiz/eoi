# Matplotlib

**Matplotlib** es una [biblioteca](https://es.wikipedia.org/wiki/Biblioteca_(programaci%C3%B3n) "Biblioteca (programación)") para la generación de gráficos a partir de datos contenidos en listas o arrays en el lenguaje de programación [Python](https://es.wikipedia.org/wiki/Python "Python") y su extensión matemática [NumPy](https://es.wikipedia.org/wiki/NumPy "NumPy"). Proporciona una API, **pylab**, diseñada para recordar a la de [MATLAB](https://es.wikipedia.org/wiki/MATLAB "MATLAB").

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
%matplotlib inline
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

# Dos interfaces por el precio de una

### Estilo matlab

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

### Estilo pitónico orientado a objetos

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

# Ajustando el plot: Colores y estilos

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

# Representaciones mediante gráfico de dispersión

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.xlim(11, -1)
plt.ylim(1.5, -1.5);

plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
plt.plot(x, y, '-ok');
```

```python
plt.plot(x, y, '-d', color='gray',
         markersize=20, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='red',
         markeredgewidth=1)
plt.ylim(-1.2, 1.2);
```

# Dispersión usando `plt.scatter`

En este caso `plt.scatter` nos permite modificar cada uno de los puntos de la representación.

```python
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100) # 100 colores random
sizes = 1000 * rng.rand(100) # muchos tamaños randoms

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='cool') #cmap es el color map
plt.colorbar();  # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar();
```

```python
# si tenemos muuuuchos puntos podríamos usar plot en vez de scatter, que es más eficiente
```

# Barras de errores

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = np.array
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k'); # error bar eje y
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
x = np.linspace(0, 10, 50)
err = np.random.rand(2,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=err, fmt='.k');
```

# Histogramas, binnings y densidad

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)
```

```python
plt.hist(data);
```

```python
plt.hist(data, bins=30, density=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', density=True, alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("trump.jpg")
red, green, blue = img.split()

red = np.array(red)
green = np.array(green)
blue = np.array(blue)


plt.hist(red)
plt.hist(green)
plt.hist(blue);
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("python.jpg")
red, green, blue = img.split()

red = np.array(red)
green = np.array(green)
blue = np.array(blue)


kwargs = dict(density=True, bins=15, alpha=0.3)

plt.hist(red, **kwargs)
plt.hist(green, **kwargs)
plt.hist(blue, **kwargs);
```
